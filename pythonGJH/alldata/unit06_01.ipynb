```python
import csv
from matplotlib import pyplot as plt
f=open('seoul.csv')
data=csv.reader(f)
next(data)
result=[]
for row in data:
    if row[-1]!='':
        result.append(float(row[-1]))
plt.hist(result,color='r',bins=100)
plt.show()
```

```python
import csv
from matplotlib import pyplot as plt
f=open('seoul.csv')
data=csv.reader(f)
next(data)
aug=[]
jan=[]
for row in data:
    if row[-1]!='':
        month=row[0].split('-')[1]
        if month=='08':
            aug.append(float(row[-1]))
        elif month=='01':
            jan.append(float(row[-1]))
plt.hist(aug,bins=100,color='r',label='Aug')
plt.hist(jan,bins=100,color='b',label='Jan')
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
import csv
f=open('seoul.csv')
data=csv.reader(f)
next(data)
day=[]
for i in range(31):
    day.append([])
for row in data:
    if row[-1]!='':
        if row[0].split('-')[1]=='08':
            day(int(row[0].split('-')[2])-1).append(float(row[-1]))
plt.style.use('ggplot')
plt.figure(figsize=((10,5),dpi=300)
plt.boxplot(day,showfliers=False)
plt.show()
```
